### 1. Calcula la distancia total (Usando integrales definidas.)  
### 2.Modelar rutas

```python
# Librerías
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, integrate, sqrt, lambdify, sin
```

```python
# Definir las funciones de las rutas
# # Ruta A: f(x) = 2x^2 - 3x + 1
# Ruta B: g(x) = sin(x) + 2
x = symbols('x')  # Variable simbólica para las funciones
f = 2*x**2 - 3*x + 1  # Ruta A
g = sin(x) + 2  # Ruta B
```

```python
# Calcular las distancias totales (longitud de arco)
distancia_f = integrate(sqrt(1 + diff(f, x)**2), (x, 0, 2))
distancia_g = integrate(sqrt(1 + diff(g, x)**2), (x, 0, 2))

print("Distancia total Ruta A:", distancia_f)
print("Distancia total Ruta B:", distancia_g)
```

```python
# ***Grafica de las rutas*** 
f_lambdified = lambdify(x, f, 'numpy')
g_lambdified = lambdify(x, g, 'numpy')

x_vals = np.linspace(0, 2, 100)
y_vals_f = f_lambdified(x_vals)
y_vals_g = g_lambdified(x_vals)  # Usar la función lambdificada

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_f, label='Ruta A: f(x) = 2x^2 - 3x + 1', color='blue')
plt.plot(x_vals, y_vals_g, label='Ruta B: g(x) = sin(x) + 2', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de Rutas')
plt.legend()
plt.grid()
plt.show()
```

```python
# Conclusiones
print("\n--- Conclusiones ---")
if distancia_f < distancia_g:
    print("La Ruta A es más eficiente en términos de distancia total.")
else:
    print("La Ruta B es más eficiente en términos de distancia total.")
```
